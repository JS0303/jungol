# 152 : 배열1 - 형성평가3
스크린샷 2022-06-06 오후 6.10.08.png
출처 : 자기주도C언어프로그래밍

```python
li = list(map(int, input().split()))

sum1 = sum2 = 0

for i in range(len(li)):
    if i % 2 == 0: # i는 짝수이지만 홀수번째로 입력받은 숫자
        sum1 += li[i]
    else:           # i는 홀수이지만 짝수번째로 입력받은 숫자
        sum2 += li[i]

print(f"odd : {sum1}")
print(f"even : {sum2}")
```

# 153 : 배열1 - 형성평가4
스크린샷 2022-06-06 오후 6.19.00.png
출처 : 자기주도C언어프로그래밍

```python
li = list(map(int, input().split()))

li2 = []

for i in li:
    if i == -1:
        break
    li2.append(i)

print(li2[-3], li2[-2], li2[-1])
```

-1이 입력되었을 때 for문을 벗어나고 그렇지 않을 시 li2라는 list에 추가해줍니다.  
그런다음 마지막 세개의 숫자를 출력하기위해 음수 인덱싱을 해주면 출력예시와 같은 결과가 나옵니다.  
그러나 이대로 제출하면  
스크린샷 2022-06-06 오후 6.33.19.png  
위와 같은 결과가 나옵니다.  
문제를 다시 살펴보면  
**"(입력받은 정수가 -1을 제외하고 3개 미만일 경우에는 -1을 제외하고 입력받은 정수를 모두 출력한다.)"**  
이런 부분이 있는데 해당 부분이 구현되지 않았으므로 해당 부분을 추가해줍니다.

```python
li = list(map(int, input().split()))

li2 = []

for i in li:
    if i == -1:
        break
    li2.append(i)

if len(li2) >= 3:
    print(li2[-3], li2[-2], li2[-1])
else:
    print(*li2)
```
