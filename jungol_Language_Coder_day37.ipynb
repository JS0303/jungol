그런다음 확인을 해보면 위와 같이 중첩 리스트의 형태로 값을 잘 받아줍니다.

```python
print('first array')
li1 = []

for _ in range(2):
    li1.append(list(map(int, input().split())))

print(li1)

print('second array')
li2 = []

for _ in range(2):
    li2.append(list(map(int, input().split())))

print(li2)
```

다음으로 2번째 배열을 입력받을 수 있도록 같은 내용의 코드를 작성합니다.

```python
print('first array')
li1 = []

for _ in range(2):
    li1.append(list(map(int, input().split())))


print('second array')
li2 = []

for _ in range(2):
    li2.append(list(map(int, input().split())))

for i in range(2): # 0,1
    for j in range(4): # 0,1,2,3
        print(li1[i][j]*li2[i][j], end=' ')
    print()
```

# 569 : 배열2 - 자가진단6
스크린샷 2022-06-12 오후 6.24.28.png
출처 : 자기주도C언어프로그래밍

```python
# 85 67 95 65
# 80 95 86 56
# 100 98 67 86
# 95 76 84 65
# 67 86 90 76
score = []

for _ in range(5):
    score.append(list(map(int, input().split())))

count = 0

for i in score:
    avg = sum(i)/len(i)
    if avg >= 80:
        count += 1
        print('pass')
    else:
        print('fail')

print(f'Successful : {count}')
```

# 570 : 배열2 - 자가진단7
스크린샷 2022-06-12 오후 6.32.14.png
출처 : 자기주도C언어프로그래밍

```python
#li = [1,1,1,1,1]

li = [1 for _ in range(5)]
print(*li)

for _ in range(4):
    for i in range(1,len(li)):
        li[i] = li[i] + li[i-1]
    print(*li)

```
