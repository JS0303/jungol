```python
a, b = map(int, input().split())
li = [a,b]

for i in range(8): # 0,1,2,...7
    li.append((li[i]+li[i+1])%10)
    
print(*li)
```

# 163 : 배열2 - 형성평가4
스크린샷 2022-06-12 오후 7.10.02.png
출처 : 자기주도C언어프로그래밍

```python
li = [[3, 5, 9], [2, 11, 5], [8, 30, 10], [22, 5, 1]] # 문제가 c언어를 기준으로 출제되어서 중괄호를 대괄호로 바꿔줘야합니다.
total = 0
# 중첩 for 문을 통한 방법
for i in li:
    for j in i:
        print(j, end=' ')
        total += j
    print()

print(total)
```

```python
li = [[3, 5, 9], [2, 11, 5], [8, 30, 10], [22, 5, 1]] # 문제가 c언어를 기준으로 출제되어서 중괄호를 대괄호로 바꿔줘야합니다.
total = 0

# for 문을 한번만 사용하는 방법
for i in li:
    print(*i)
    total += sum(i)

print(total)
```
