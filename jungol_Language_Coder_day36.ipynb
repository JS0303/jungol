# 566 : 배열2 - 자가진단3
스크린샷 2022-06-12 오후 5.54.00.png  
출처 : 자기주도C언어프로그래밍

```python
n = int(input())

li = [100, n]
i = 1

# 조건에 따라 몇개의 숫자가 나올지 정확히 모르므로 while문 사용
while li[i] >= 0:
    li.append(li[i-1] - li[i])
    i += 1

print(*li)
```

# 567 : 배열2 - 자가진단4
스크린샷 2022-06-12 오후 5.59.58.png  
출처 : 자기주도C언어프로그래밍

```python
# 풀이1
li = [[5,8,10,6,4],[11,20,1,13,2],[7,9,14,22,3]]

# 중첩 for문을 통해 해결
for i in range(len(li)): # 0,1,2
    for j in range(len(li[i])): # 0,1,2,3,4
        print(f'{li[i][j]:>2}', end='   ')
    print()
```

```python
# 풀이2
li = [[5,8,10,6,4],[11,20,1,13,2],[7,9,14,22,3]]

for i in li:
    for j in i:
        print(f'{j:>2}', end='   ')
    print()
```
