# 573 : 함수1 - 자가진단3
스크린샷 2022-06-28 오후 11.08.13.png
출처 : 자기주도C언어프로그래밍

```python
def number_square(n):
    for i in range(n): # 0,1,2,3
        for j in range(1, n+1): # 1,2,3,4
            print(n*i+j, end=' ')
        print()

n = int(input())

number_square(n)
```

# 574 : 함수1 - 자가진단4
스크린샷 2022-06-28 오후 11.14.50.png
출처 : 자기주도C언어프로그래밍

```python
def maximum(a, b, c):
    return max(a,b,c)

a, b, c = map(int, input().split())

print(maximum(a, b, c))
```
