```python
li = list(map(int, input().split()))

sum_ = 0 # 합을 담을 변수
count_ = 0 # 개수를 담을 변수

for i in li:
    if i == 0: # 0이 입력될 경우
        break
    if i % 5 == 0: # 5의 배수일 경우
        sum_ += i
        count_ += 1

print(f"Multiples of 5 : {count_}")
print(f"sum : {sum_}")
print(f"avg : {sum_/count_:.1f}")
```

python으로 작성시 hint가 있어 정리합니다.  
입력받은 값이 0이 아닐 경우 result[ ]에 담습니다.  
0이 나오면 result[ ]를 반환하는 함수를 작성하라고 합니다.  
hint에 제시된 내용을 바탕으로 함수의 빈 부분을 채우면 될것같습니다.  

```python
def getPositiveNumber(alist): 
# ADD ADDITIONAL CODE HEAR !!!
    result = []
    for i in alist:
        if i == 0:
            break
        result.append(i)

    return result
 
alist = list(map(int, input().split()))
alist = getPositiveNumber(alist) 
print(alist)
# for i in alist: 
#     if i% 2 == 0: 
#         print(i//2, end = ' ') 
#     else: 
#         print(i*2, end = ' ')
```

함수의 빈 부분을 채운 뒤 alist를 출력해보면 0이 입력되기 이전의 값들만 들어가있음을 알 수 있습니다.

```python
def getPositiveNumber(alist): 
# ADD ADDITIONAL CODE HEAR !!!
    result = []
    for i in alist:
        if i == 0:
            break
        result.append(i)

    return result
 
alist = list(map(int, input().split()))
alist = getPositiveNumber(alist) 

# 짝수일 경우 나누기 2, 홀수일 경우 곱하기 2
# 코드가 다 작성되어있습니다.
for i in alist: 
    if i% 2 == 0: 
        print(i//2, end = ' ') 
    else: 
        print(i*2, end = ' ')
```

여기까지 작성하면 list의 값은 정상적으로 출력됩니다.  
하지만 여기서 한가지 추가해줘야하는데 입력받은 갯수가 몇개인지 입니다.  
따라서 해당 코드를 한줄 추가해줍니다.

```python
def getPositiveNumber(alist): 
# ADD ADDITIONAL CODE HEAR !!!
    result = []
    for i in alist:
        if i == 0:
            break
        result.append(i)

    return result
 
alist = list(map(int, input().split()))
alist = getPositiveNumber(alist) 

print(len(alist))

# 짝수일 경우 나누기 2, 홀수일 경우 곱하기 2
# 코드가 다 작성되어있습니다.
for i in alist: 
    if i% 2 == 0: 
        print(i//2, end = ' ') 
    else: 
        print(i*2, end = ' ')
```
