# 576 : 함수1 - 자가진단6
스크린샷 2022-06-29 오후 11.00.17.png
출처 : 자기주도C언어프로그래밍

```python
def calculation(x, ch, y ):
    if ch == '+':
        print(f'{x} {ch} {y} = {x+y}')
    elif ch == '-':
        print(f'{x} {ch} {y} = {x-y}')
    elif ch == '*':
        print(f'{x} {ch} {y} = {x*y}')
    elif ch == '/':
        print(f'{x} {ch} {y} = {x//y}')
    else:
        print(f'{x} {ch} {y} = 0')    

x, ch, y = input().split()
x = int(x)
y = int(y) 
calculation(x, ch, y)
```

# 577 : 함수1 - 자가진단7
스크린샷 2022-06-29 오후 11.06.25.png
출처 : 자기주도C언어프로그래밍

```python
def number(a,b):
    if a>b:
        a = a//2
        b = b*2
    else:
        a = a*2
        b = b//2
    return a, b

a,b = map(int, input().split())
a,b = number(a,b)
print(a,b)
```

# 578 : 함수1 - 자가진단8
스크린샷 2022-06-29 오후 11.10.44.png
출처 : 자기주도C언어프로그래밍

```python
def googoodan(n):
    print(f'== {n}dan ==')
    for i in range(1, 10):
        print(f'{n} * {i} = {n*i:>2}')
    print()

a,b = map(int,input().split())

for n in range(min(a,b),max(a,b)+1):
    googoodan(n)
```
