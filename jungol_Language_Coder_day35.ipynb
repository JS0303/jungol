# 564 : 배열2 - 자가진단1
스크린샷 2022-06-07 오후 11.14.22.png
출처 : 자기주도C언어프로그래밍


```python
li = input().split()

text = []

for i in li:
    if i.isupper():
        text.append(i)

for i in range(65, 91): # 알파벳 A의 아스키코드 : 65, 알파벳 Z의 아스키코드 : 90
    ch = chr(i)
    if ch in text:
        print(f"{ch} : {text.count(ch)}")
```

# 565 : 배열2 - 자가진단2
스크린샷 2022-06-07 오후 11.23.32.png
출처 : 자기주도C언어프로그래밍

```python
li = list(map(int, input().split()))
num = []

for i in li:
    if i == 0:
        break
    num.append((i%100)//10) # 10의자리 숫자만 남겨서 list에 추가

for i in range(10):
    if i in num:
        print(f"{i} : {num.count(i)}")
```
