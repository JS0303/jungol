# 562 : 배열1 - 자가진단8
스크린샷 2022-06-04 오후 5.55.33.png

출처 : 자기주도C언어프로그래밍

```python
li = list(map(int, input().split()))
# 입력받는 순서를 짝수, 홀수로 구분하고 짝수번째 입력의 합은 sum1에, 홀수번째의 합은 sum2에 담아주겠습니다. 
sum1 = sum2 = 0
for i in range(len(li)):
    if i % 2 == 0: # i가 짝수라면 홀수번째 입력(index는 0부터 시작하므로)
        sum2 += li[i]
    else:           # i가 홀수라면 짝수번째 입력(index는 0부터 시작하므로) 
        sum1 += li[i]

print(f"sum : {sum1}")
print(f"avg : {sum2/5:.1f}") # 10개의 숫자를 입력받았으므로 5로 나누어줍니다.
```

출처 : 자기주도C언어프로그래밍

```python
li = list(map(int, input().split()))

li.sort(reverse=True) # sort 메소드는 기본적으로 오름차순 정렬이므로 reverse=True 설정을 주어 내림차순으로 바꿔줍니다.

print(*li)
```
