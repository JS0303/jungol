```python
# dictionary를 사용한 방법 (key : value)
li = list(map(int, input().split()))
dic = {key:0 for key in range(1,7)}

for i in li:
    dic[i] += 1

for key, value in dic.items():
    print(f'{key} : {value}')
```

```python
scores = list(map(int, input().split()))
person = {key:0 for key in range(100,-1,-10)} # 100부터 0까지

for i in scores:
    if i == 0:
        break
    if i == 100:
        person[100] += 1
    elif i == 90:
        person[90] += 1

print(person)
```

```python
scores = list(map(int, input().split()))
person = {key:0 for key in range(100,-1,-10)} # 100부터 0까지

for i in scores:
    if i == 0:
        break
    if i == 100:
        person[100] += 1
    elif i >= 90:
        person[90] += 1
    elif i >= 80:
        person[80] += 1
    elif i >= 70:
        person[70] += 1
    elif i >= 60:
        person[60] += 1
    elif i >= 50:
        person[50] += 1
    elif i >= 40:
        person[40] += 1
    elif i >= 30:
        person[30] += 1
    elif i >= 20:
        person[20] += 1
    elif i >= 10:
        person[10] += 1
    else:
        person[0] += 1

for key, value in person.items():
    if value != 0:
        print(f'{key} : {value} person')
```
