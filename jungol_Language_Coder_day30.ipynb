```python
li = list(map(int, input().split()))

min_ = li[0]
max_ = li[0]

for i in li:
    if i < min_:
        min_ = i
    if i > max_:
        max_ = i

print(min_, max_)
```

우선 최소값과 최대값을 담을 변수를 하나씩 만들고 for문을 통해 찾아줍니다.

```python
li = list(map(int, input().split()))

min_ = li[0]
max_ = li[0]

for i in li:
    if i < min_ and i >= 100:
        min_ = i
    if i > max_ and i < 100:
        max_ = i

print(max_, min_)
```

그런 다음 주어진 조건에 따라 최소값의 경우 100보다 크거나 같고 최대값의 경우 100보다 작다는 조건을 추가해줍니다.  
그러나 결과값은 정답과 다른데 이유는 최소값을 초기화할 때 맨 앞 인덱스의 값이 들어가도록 설정이 되어있기 때문입니다.  
그러므로 초기값을 수정해주어야 하는데 hint를 참고하면

위와 같은 내용을 확인할 수 있습니다.  
따라서 최소값은 10000, 최대값은 0으로 설정해줍니다.

```python
li = list(map(int, input().split()))

min_ = 10000
max_ = 0

for i in li:
    if i < min_ and i >= 100:
        min_ = i
    if i > max_ and i < 100:
        max_ = i

print(max_, min_)
```

여기서 제출하면 60점이라는 점수가 나오는데  
hint의 내용을 확인해보면 해당하는 수가 없을 경우 100을 출력한다고 되어있습니다.  
따라서 해당 조건을 추가해줍니다.


```python
# ex) 1부터 10까지의 숫자를 넣었을 때
li = list(map(int, input().split()))

min_ = 10000
max_ = 0

for i in li:
    if i < min_ and i >= 100:
        min_ = i
    if i > max_ and i < 100:
        max_ = i

print(max_, min_)
```

```python
# 조건을 추가한 코드
li = list(map(int, input().split()))

min_ = 10000
max_ = 0

for i in li:
    if i < min_ and i >= 100:
        min_ = i
    if i > max_ and i < 100:
        max_ = i

if min_ == 10000:
    min_ = 100
if max_ == 0:
    max_ = 100

print(max_, min_)
```
