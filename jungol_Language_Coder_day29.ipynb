# 558 : 배열1 - 자가진단4
스크린샷 2022-06-04 오후 3.51.01.png

출처 : 자기주도C언어프로그래밍

```python
li = list(map(int, input().split()))
li2 = []

for i in li:
    if i == 0:
        break
    li2.append(i)
li2.reverse()

print(*li2)
```
