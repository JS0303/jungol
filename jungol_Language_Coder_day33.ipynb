# 156 : 배열1 - 형성평가7
스크린샷 2022-06-06 오후 7.02.31.png
출처 : 자기주도C언어프로그래밍

```python
li = list(map(int, input().split()))
li2 = []

for i in li:
    if i == 999:
        break
    li2.append(i)

print(li2)
```

li 안에 값들을 입력받고 li2라는 빈 list를 생성해줍니다.  
for문을 통해 999라는 값이 입력된다면 멈춰주고 그렇지 않다면 li2에 추가해줍니다.  
이상태로 예시를 입력해 결과를 출력하면 999이전까지 입력된 값들만 출력됩니다.  
이제 최소값과 최대값을 찾아 출력해주면 됩니다.

```python
li = list(map(int, input().split()))
li2 = []

for i in li:
    if i == 999:
        break
    li2.append(i)

print(f"max : {max(li2)}")
print(f"min : {min(li2)}")
```

max와 min함수를 사용하면 f string을 통해 문제에서 요구하는 최대값과 최소값을 간단하게 찾을 수 있습니다.  
여기서 주의할 점은 max와 min함수 안에 원본 li가 아닌 새롭게 생성된 li2가 들어가야 한다는점 입니다.
